```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from collections import Counter

text = 'Life is a bitch and a bitch is life'
words = text.split()

# Start with one symbol per character.
initial_symbols = sorted(set(''.join(words)))
vocab = {ch: i for i, ch in enumerate(initial_symbols)}

print('Initial vocabulary:', vocab)
```

```python
def byte_pair_encode(words, num_merges=5):
    tokenized_words = [list(word) for word in words]
    vocab = {ch: i for i, ch in enumerate(sorted(set(''.join(words))))}
    merges = []

    for _ in range(num_merges):
        pair_counts = Counter()

        for tokenized_word in tokenized_words:
            for i in range(len(tokenized_word) - 1):
                pair_counts[tuple(tokenized_word[i:i + 2])] += 1

        if not pair_counts:
            break

        pair, _ = max(pair_counts.items(), key=lambda item: (item[1], item[0]))
        new_symbol = pair[0] + pair[1]

        merged_words = []
        for tokenized_word in tokenized_words:
            merged = []
            i = 0
            while i < len(tokenized_word):
                if i < len(tokenized_word) - 1 and tuple(tokenized_word[i:i + 2]) == pair:
                    merged.append(new_symbol)
                    i += 2
                else:
                    merged.append(tokenized_word[i])
                    i += 1
            merged_words.append(merged)

        tokenized_words = merged_words
        vocab[new_symbol] = max(vocab.values()) + 1
        merges.append((pair, new_symbol))

    return tokenized_words, vocab, merges


merged_words, vocab, merges = byte_pair_encode(words, num_merges=5)
merged_words
```

```python
for pair, new_symbol in merges:
    print(f'merged {pair} -> {new_symbol}')

print('\nFinal vocabulary:')
for symbol, idx in sorted(vocab.items(), key=lambda item: item[1]):
    print(f'{idx}: {symbol}')

print(f'\nNumber of tokens after BPE: {sum(len(word) for word in merged_words)}')
```
