```python
#we need to create a function which is a encoder and a decoder
vocab = set()
for sentence in sentences:
    for word in sentence.lower().split():
        vocab.add(word)

vocab = sorted(vocab)  # sort for consistent indexing
```

```python
word2idx = {}
for i , word in enumerate(vocab):
    word2idx[word] = i
word2idx
```

```python
idx2word = {idx: word for word, idx in word2idx.items()}
print(len(idx2word))
```

```python
def encoder(sentences):
    return [word2idx[word] for word in sentence.lower().split() if word in word2idx]
```

```python
def decoder(indices):
    return "".join([idx2word[i] for i in indices])
```

```python
# Test
sample = "Levi moves faster than any soldier"
encoded = encoder(sample)
decoded = decoder(encoded)

print("Encoded:", encoded)
print("Decoded:", decoded)
```
